# Description

> The `following` description is based on *Murphy, Kevin P.: 'Machine learning: a propabilistic perspective' pp. 472*.

## Aim

We have data from a population that follows an univariate Normal distribution

$$
\begin{align}
X \sim N(\mu, \lambda^{-1})
\end{align}
$$

with unknown parameters $\mu$, $\lambda$, which we aim to infer. We use a [normal-gamma distribution](http://en.wikipedia.org/wiki/Normal-gamma_distribution) as prior for $\mu$ and $\lambda$:

$$P(\mu, \lambda)=N(\mu|\mu_0, (\kappa_0\lambda)^{-1})\Gamma(\lambda|a_0, b_0)$$

The hyperparameters $\mu_0$, $\kappa_0$, $a_0$, and $b_0$ capture our prior knowledge about $\mu$ and $\lambda$:

  * $E[\mu]=u_0$, $Var[\mu]=\frac{b_0}{\kappa_0(a_0-1)}$

  * $E[\lambda]=\frac{a_0}{b_0}$, $Var[\lambda]=\frac{a_0}{b_0^2}$
    
Given a sample $D=\{x_1,...,x_N\}$, the posterior distribution $P(\mu,\lambda|D)$ is:

$$P(\mu, \lambda|D) \propto P(\mu, \lambda, D) = P(D|\mu,\lambda) P(\mu, \lambda) = \prod_{i=1}^N N(x_i|\mu, \lambda^{-1}) N(\mu|\mu_0, (\kappa_0\lambda)^{-1})\Gamma(\lambda|a_0, b_0)$$

Since our normal-gamma prior is conjugate to a normal distribution, the posterior $P(\mu,\lambda|D)$ follows itself a normal-gamma distribution ([proof](http://en.wikipedia.org/wiki/Normal-Gamma#Posterior_distribution_of_the_parameters)), and thus allows effiecent inference of $\mu$ and $\lambda$. Yet, we will use [variational inference](http://johnwinn.org/Publications/papers/VMP2005.pdf) to approximate $P(\mu, \lambda|D)$ by a simpler distribution $Q(\mu, \lambda)$. 

## Variational inference

The [Kullback-Leibler divergence](http://en.wikipedia.org/wiki/Kullback%E2%80%93Leibler_divergence) $D(P(x)||Q(x))$ quantifies the divergence of an approximate distribution $Q(x)$ from the real distribution $P(x)$:

$$D(P(x)||Q(x))=\int_x P(x) \log \frac{P(x)}{Q(x)}$$

Variational inference uses the *reverse Kullback-Leiber divergence* $D(Q(x)||P(x))$, to find a tractable surrogate $Q(x)$ for an untractable distribution $P(x)$:

$$D(Q(x)||P(x)) = D(Q(x)||\tilde{P}(x)) + \log Z$$

$\log Z$ is the intractable normalization constant for the unnormalized distribution $\tilde{P}(x)$. Since it does not depend on $Q(x)$, we can use following cost function to measure the fit between $Q(x)$ and $P(x)$:

$$J(Q(x)) = D(Q(x)||\tilde{P}(x)) = D(Q(x)||P(x)) - \log Z \propto D(Q(x)||P(x))$$

Since $D(Q(x)||P(x))>0$, $J(Q(x))$ is an upper bound for $-\log Z$. 

Let $Q_i(x)$ be the factors of $Q(x)$, i.e. $Q(x)=\prod_i Q_i(x)$. Then it can be shown that $J(Q(x))$ is minimal, if

$\log Q_i(x) \propto E_{-Q_i}[\log \tilde{P}(x)]$,

where $E_{-Q_i}[f(x)]$ is the expectation of $f(x)$ using the density $Q(x)$ but the factor $Q_i(x)$.

## Variational inference for an univariate normal

We want to approximate $P(\mu,\lambda|D)$ by $Q(\mu, \lambda)$, which shall factorize as follows:

$$Q(\mu, \lambda) = Q_\mu(\mu) Q_\lambda(\lambda)$$

The cost function is then:

$$J(Q(\mu,\lambda)) = D(Q(\mu,\lambda)||P(\mu,\lambda, D))$$

$J(Q(\mu,\lambda))$ is an upper bound for the negative log likelihood $-\log P(D)$. 

For deriving $Q_\mu(\mu)$ and $Q_\lambda(\lambda)$, we need to compute the expectation over $\log P(\mu,\lambda, D)$:

\begin{align}
\log P(\mu,\lambda, D) &= \log P(D|\mu, \lambda) + \log P(\mu|\lambda) + \log P(\lambda) = \\
  &\propto \frac{N}{2}\log \lambda - \frac{\lambda}{2}\sum_{i=1}^N (x_i-\mu)^2 - \frac{\kappa_0\lambda}{2}(\mu-\mu_0)^2 \\
  &\quad + \frac{1}{2}\log(\kappa_0\lambda)+(a_0-1)\log\lambda - b_0\lambda
\end{align}

By solving $\log Q_\mu(\mu)=E_{Q_\lambda}[\log P(\mu,\lambda, D)]$, it can be shown that

$$Q_\mu(\mu) = N(\mu|\mu_N, \kappa_N^{-1})$$

with parameters:

$$\mu_n = \frac{\kappa_0\mu_0 + N \bar{x}}{\kappa_0+N}$$
$$\kappa_N = (\kappa_0 + N)\frac{a_N}{b_N}$$

By solving $\log Q_\lambda(\lambda)=E_{Q_\mu}[\log P(\mu,\lambda, D)]$, it can be shown that

$$Q_\lambda(\lambda) = \text{Ga}(\lambda|a_N, b_N)$$ 

with parameters:

$$a_N = a_0 + \frac{N+1}{2}$$
$$b_N = b_0 + \kappa_0(\frac{1}{\kappa_N}+\mu_N^2 + \mu_0^2 - 2\mu_N\mu_0) + \frac{1}{2}\sum_{i=1}^N(x_i^2+\frac{1}{\kappa_N}+\mu_N^2-2\mu_N x_i)$$

Note that $\mu_0$, $\kappa_0$, $a_0$, and $b_0$ are the hyperparameters of the prior $P(\mu,\lambda)$, whereas $\mu_N$, $\kappa_N$ are the parameters of $Q_\mu(\mu)$, and $a_N$, $b_N$ of $Q_\lambda(\lambda)$. By initializing $\mu_N$, $\kappa_N$, as well as $a_N$, $b_N$, and recursively updating them, $Q(\mu,\lambda)$ will approximate $P(\mu,\lambda|D)$.


# Simulations

## Source code

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import scipy.stats as stats
import sys

sys.path.append('..')
from varinf import uninorm

%matplotlib inline
```

```python
def plot_posterior(posterior, axes, xlim=None, ylim=None, color='red'):
    if xlim is None or ylim is None:
        lower = posterior.mean() - 3 * posterior.var()
        upper = posterior.mean() + 3 * posterior.var()
        if xlim is None:
            xlim = [lower[0], upper[0]]
        if ylim is None:
            ylim = [lower[1], upper[1]]

    locs = np.linspace(xlim[0], xlim[1])
    scales = np.linspace(ylim[0], ylim[1])
    pp = posterior.pdf(locs, scales)
    axes.contour(locs, scales, pp, colors=color)
    axes.set_xlabel('$\mu$')
    axes.set_ylabel('$\lambda$')
```

```python
def plot_infer_results(pposterior, qposterior_infer_results, maxit=None, xlim=None, ylim=None):
    opt_index = qposterior_infer_results.opt_index()
    ncols = 3
    if maxit is None:
        nrows = int(np.ceil(float(opt_index) / ncols))
        maxit = nrows * ncols + 1
    else:
        nrows = int(np.ceil(float(maxit + 1) / ncols))
         
    fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(ncols * 5, nrows*5))
    axes = axes.flatten()
    for i in range(0, maxit + 1):
        plot_posterior(pposterior, axes[i], xlim, ylim, color='red')
        qposterior = uninorm.QPosterior(qposterior_infer_results[i])
        plot_posterior(qposterior, axes[i], xlim, ylim, color='green')
        axes[i].set_title('Iteration {:d}{:s}'.format(i, '*' if i == opt_index else ''))   
```

```python
def plot_costs(results):
    fig, axes = plt.subplots(figsize=(15, 5))
    line, = axes.plot(range(1, len(results)), results.c[1:])
    line.set_linewidth(2)
    line.set_marker('s')
    line.set_markerfacecolor('red')
    line.set_color('blue')
    axes.set_xlabel('Iteration')
    axes.set_ylabel(r'$-J(Q(\mu,\lambda))$')
    axes.set_xticks(range(1, len(results)))
    axes.grid()
```

## Setting 1

```python
np.random.seed(0)
# 10 points sampled from N(u=0.0, l=1.0)
x = uninorm.simulate(u=0.0, l=1.0, size=10)
# Prior: E[u]=0.0 (weak), E[l]=1.0 (weak)
pprior_params = uninorm.Params(u=0.0, k=1.0, a=0.001, b=0.001)
pposterior_params = uninorm.infer_pposterior(x, pprior_params)
pposterior = uninorm.PPosterior(pposterior_params)
init = uninorm.Params(u=0.0, k=1.0, a=2.0, b=1.0)
qposterior_infer_results = uninorm.infer_qposterior(x, pprior_params, init, maxit=10)

plot_infer_results(pposterior, qposterior_infer_results, maxit=5, xlim=[-0.5, 1.8], ylim=[0.01, 2.5])
plot_costs(qposterior_infer_results)
print(qposterior_infer_results)
```

# Aside: Gamma distribution

```python
gamma_as = [1.0, 2.0, 5.0]
gamma_bs = [0.5, 1.0, 2.0]
xrange = np.linspace(0, 10)
fig, axes = plt.subplots(nrows=1, ncols=len(gamma_as), figsize=(len(gamma_as) * 5, 4))
for i, axis in enumerate(axes):
    for gamma_b in gamma_bs:
        gamma = stats.gamma(a=gamma_as[i], scale=1.0/gamma_b)
        yrange = gamma.pdf(xrange)
        line, = axis.plot(xrange, yrange)
        line.set_linewidth(2)
        line.set_label('$a={:.1f}$  $b={:.1f}$'.format(gamma_as[i], gamma_b))
    axis.grid()
    axis.legend(loc=1)
```
